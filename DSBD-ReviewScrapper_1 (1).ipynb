```python
soup = bs4.BeautifulSoup(requests1.text)
```

```python
reviews = soup.findAll('div',{'class':'t-ZTKy'})
for review in reviews:
    print(review.get_text()+"\n\n")
```

```python
ratings = soup.find('div',{'class':'_1azcI6'}).get_text();
print(ratings)
```

```python
ratings = soup.find_all('div',{'class':'_3LWZlK _1BLPMq'})
for rating in ratings:
    print(rating.get_text()+"\n")
```

```python
tags = soup.find('span',{'class':'_1V_ZGU'}).get_text();
tags
```

```python
customer_name = soup.findAll('p',{'class':'_2sc7ZR _2V5EHH'})
for cust_name in customer_name:
    print(cust_name.get_text()+" \n")
```

```python
price = soup.find('div',{'class':'_30jeq3 _16Jk6d'}).get_text();
print(price)
```

```python
questions = soup.findAll('div',{'class':'_1xR0kG _3cziW5'})
for question in questions:
    print(question.get_text())
    answer = question.findNext('div',{'class':'_2yeNfb'}).get_text();
    print(answer+"\n\n")
```
